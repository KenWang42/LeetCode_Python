[Pre Order Traversal](https://leetcode.com/explore/learn/card/data-structure-tree/134/traverse-a-tree/928/)

V-L-R

Itterative用stack實現

```
# Recursively
class Solution:
    def __init__(self):
        self.res=[]
        
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        if(root):
            self.res.append(root.val)
            self.preorderTraversal(root.left)
            self.preorderTraversal(root.right)
        return self.res
    
# Iteratively
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            if node:
                res.append(node.val)
                stack.append(node.right)
                stack.append(node.left)
        return res
```

[In Order Traversal](httpshttps://leetcode.com/explore/learn/card/data-structure-tree/134/traverse-a-tree/929/)

L-V-R

Recursive is faster

```
# Recursively
class Solution:
    def __init__(self):
        self.res=[]
        
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        if(root):
            self.inorderTraversal(root.left)
            self.res.append(root.val)
            self.inorderTraversal(root.right)
        return self.res

# Iteratively
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        res = []
        stack = []
        while True:
            while root:
                stack.append(root)
                root = root.left
            if not stack:
                return res
            node = stack.pop()
            res.append(node.val)
            root = node.right
```

[Post Order Traversal](https://leetcode.com/explore/learn/card/data-structure-tree/134/traverse-a-tree/930/)

L-R-V

```
# Recursively
class Solution:
    def __init__(self):
        self.res=[]
    
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        if(root):
            self.postorderTraversal(root.left)
            self.postorderTraversal(root.right)
            self.res.append(root.val)
        return self.res

# Iteratively
class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        if not root:
            return []
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            res.append(node.val)
            if node.left:
                stack.append(node.left)
            if node.right:
                stack.append(node.right)
        return res[::-1] ## 倒著輸出
```

[Level Order Traversal](https://leetcode.com/explore/learn/card/data-structure-tree/134/traverse-a-tree/931/)

順著層數 從左到右

```
# Iteratively
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        if not root:
            return []
        
        queue = [root]
        res = []
        
        while queue:
            temp = []
            for i in range(len(queue)):
                node = queue.pop(0)
                temp.append(node.val)
                
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                
            res.append(temp)
        return res
```

[Maximum Depth of Binary Tree](https://leetcode.com/explore/learn/card/data-structure-tree/17/solve-problems-recursively/535/)

Itteratively: BFS

```
# Recursively
class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        if not root: return 0
        # Bottom-up
        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)
        return max(left_depth, right_depth) + 1
    
# One Line
    def maxDepth(self, root):
        return 1 + max(map(self.maxDepth, (root.left, root.right))) if root else 0

# Iteratively
    def maxDepth(self, root):
        depth = 0
        level = [root] if root else []
        while level:
            depth += 1
            next_level = []
            for node in level:
                if node.left:
                    next_level.append(node.left)
                if node.right:
                    next_level.append(node.right)
            level = next_level
        return depth
```

[Symmetric Tree](https://leetcode.com/explore/learn/card/data-structure-tree/17/solve-problems-recursively/536/)

判斷樹是不是對稱的

對稱: 從每個點看下去 左右的數值一樣 或是 都沒有

```
    # Recursively
    def isSymmetric(self, root: TreeNode) -> bool:
        if not root: return True
        else: return self.isSym(root.left, root.right)
        
    def isSym(self, left, right) -> bool:
        if (not left) and (not right): return True ## 兩邊都沒有
        if (not left) or (not right): return False ## 其中一個沒有
        if left.val == right.val:  ## 兩邊值一樣
            out_pair = self.isSym(left.left, right.right)
            in_pair = self.isSym(left.right, right.left)
            return out_pair and in_pair
        return False
    
    # Iteratively: Stack
    def isSymmetric(self, root: TreeNode) -> bool:
        if not root: return True
        stack = [[root.left, root.right]]
        while stack:
            left, right = stack.pop() ## left = root.left, right = root.right
            if (not left) and (not right):
                continue
            if (not left) or (not right):
                return False
            if left.val == right.val:
                stack.append((left.left, right.right))
                stack.append((left.right, right.left))
            else:
                return False
        return True
```

[Path Sum](https://leetcode.com/explore/learn/card/data-structure-tree/17/solve-problems-recursively/537/)

指定數字 從 root 一路加到某 leaf (沒有children的點) 有沒有相等於該數字

```
# Recursively
class Solution:
    def hasPathSum(self, root: TreeNode, sum: int) -> bool:
        if not root:
            return False
        if (not root.left) and (not root.right) and (sum - root.val):
            return True
        return self.hasPathSum(root.left, sum - root.val) or self.hasPathSum(root.right, sum - root.val)
```

Construct Binary Tree

[Post Order and In Order](https://leetcode.com/explore/learn/card/data-structure-tree/133/conclusion/942/)

Post Order 的最後一項一定是 root

在 In Order中 該 root 可以區分左右兩邊的 children

```
# Recursively
class Solution:
    def buildTree(self, inorder: List[int], postorder: List[int]) -> TreeNode:
        if not inorder or not postorder:
            return None
        while inorder:
            root = TreeNode(postorder.pop())
            inorder_index = inorder.index(root.val)
            
            root.right = self.buildTree(inorder[inorder_index + 1:], postorder)
            root.left = self.buildTree(inorder[:inorder_index], postorder)
            return root
```

Construct Binary Tree

[Pre Order and In Order](https://leetcode.com/explore/learn/card/data-structure-tree/133/conclusion/943/)

Pre Order 的第一項一定是 root

在 In Order中 該 root 可以區分左右兩邊的 children

```
# Recursively
class Solution:
     def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        if not inorder or not preorder:
             return None
        while inorder:
            root = TreeNode(preorder.pop(0))
            inorder_index = inorder.index(root.val)
            
            root.left = self.buildTree(preorder, inorder[0:inorder_index])
            root.right = self.buildTree(preorder, inorder[inorder_index+1:])
            return root
        
# Iteratively: Stack
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        if len(preorder) == 0:
            return None
        
        head = TreeNode(preorder[0])
        stack = [head]
        i = 1
        j = 0
        
        while i < len(preorder):
            temp = None
            t = TreeNode(preorder[i])
            while stack and stack[-1].val == inorder[j]:
                temp = stack.pop()
                j += 1
            if temp:
                temp.right = t
            else:
                stack[-1].left = t
            stack.append(t)
            i += 1
        return head
```

[Populating Next Right Pointers in Each Node](https://leetcode.com/problems/populating-next-right-pointers-in-each-node/)

其實就是 Level Order Traversal 改

因為是完全樹 left 一定會有 next 指到 right 上

沒有 next: right.next 就沒東西了

```
class Solution:
    def connect(self, root: 'Node') -> 'Node':
        res = root
        while root and root.left:
            next_level = root.left
            while root:
                root.left.next = root.right
                if root.next:
                    root.right.next = root.next.left
                root = root.next
            root = next_level
            
        return res
```

[ Populating Next Right Pointers in Each Node II](https://leetcode.com/explore/learn/card/data-structure-tree/133/conclusion/1016/)

跟上一題不同在於 不一定是完全樹

用 level 把下一層的東西暫存起來



```
from collections import deque
class Solution:
    def connect(self, root: 'Node') -> 'Node':
        if not root: return
        
        res = root
        
        queue, level = deque([root]), deque()
        
        while queue:
            node = queue.popleft()
            if node.left:
                level.append(node.left)
            if node.right:
                level.append(node.right)
            node.next = queue[0] if queue else None
            if not queue and level:
                queue = level
                level = deque()
        
        return res
```

[ Lowest Common Ancestor of a Binary Tree
](https://leetcode.com/explore/learn/card/data-structure-tree/133/conclusion/932/) 

 找到最近的共同祖先
 
 LCA(x, y) =
 
 { x,    if x = y or x = parent(y)
 
 { y,    if y = parent(x)
 
 { LCA(x, parent(y)),    otherwise
 
 { LCA(parent(x), y),    otherwise
 
 從 root 開始 找到 p 或是 q 就回到其祖先
 
 直到回到同樣的點 (該點會收到兩個不是null的值)


```
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if (q == root) or (p == root) or (root is None):
            return root
        right = self.lowestCommonAncestor(root.right, p, q)
        left = self.lowestCommonAncestor(root.left, p, q)
        if left and right: ## 收到兩個值了
            return root  
        else:
            return left or right ## 回傳不是 null 的項目
```
